```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件
plt.style.use('ggplot')

# 忽略警告訊息
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
"""
Your Code Here
"""
matrix = (np.random.random((10,10)) - 0.5) * 2 
plt.figure(figsize=(10,10))
"""
Your Code Here
"""
heatmap = sns.heatmap(matrix,cmap = plt.cm.RdYlBu_r,vmin = -1 ,vmax = 1,annot = True)
plt.show()
```

```python
nrow = 1000
ncol = 3
"""
Your Code Here
"""
matrix = (np.random.random((1000,3)) - 0.5 )* 2
# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size=nrow)
#以indice作為matrix的索引，並透過reset_index，將舊索引添加為一行
plot_data = pd.DataFrame(matrix, indice).reset_index()
print(plot_data)

# 繪製 seborn 進階 Heatmap
grid = sns.PairGrid(data = plot_data, height = 3, diag_sharey=False, 
                    hue = 'index', vars = [x for x in list(plot_data.columns) if x != 'index'])
#以不同index 來畫圖，所以會有三種圖
"""
Your Code Here
Please replace "..." to correct plot function
"""
grid.map_upper(plt.scatter , alpha = 0.2)
grid.map_diag(sns.kdeplot)
grid.map_lower(sns.kdeplot, cmap=plt.cm.OrRd_r)

# plt.show()
```

```python
nrow = 1000
ncol = 3
"""
Your Code Here
"""

# randn會產生隨機分布，reshape函數將一維數組重新塑形為二維數組
matrix = np.random.randn(nrow * ncol).reshape((nrow,ncol))
print(matrix)

# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice).reset_index()

# 繪製 seborn 進階 Heatmap
grid = sns.PairGrid(data = plot_data, height = 3, diag_sharey=False,
                    hue = 'index', vars = [x for x in list(plot_data.columns) if x != 'index'])
"""
Your Code Here
Please replace "..." to correct plot function
"""
grid.map_upper(plt.scatter, alpha=0.2)
grid.map_diag(sns.kdeplot)
grid.map_lower(sns.kdeplot, cmap=plt.cm.OrRd_r)

# plt.show()
```
